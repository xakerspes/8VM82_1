```python
#Вывод таблицы для просмотра данных
cyul_2018 = pd.read_csv('ideal.csv', encoding='windows-1252')
cyul_2018.set_index('DateTime', inplace=True)
cyul_2018
```

```python
#Первый период A
data3=cyul_2018['2017-09-01 00:00:00+00':'2017-09-01 10:00:00+00']
data_per = data3['Tmean']
data_per
```

```python
#Второй период B
data3_2=cyul_2018['2017-10-01 00:00:00+00':'2017-10-01 10:00:00+00']
data_per2 = data3_2['Tmean']
data_per2
```

```python
#Первый период
plt.figure(figsize=[18,10])
plt.plot(data_per)
plt.xlabel('Период', fontsize=15)
plt.ylabel('$^oC$', fontsize=15)
plt.title('Выбранное значение для первого периода', fontsize=25)
```

```python
temp_fft = sp.fftpack.fft(data_per)
temp_psd = np.abs(temp_fft) ** 2
fftfreq = sp.fftpack.fftfreq(len(temp_psd), 1)
i = fftfreq > 0
fig, ax = plt.subplots(1, 1, figsize=(18, 10))
ax.plot(fftfreq[i], 10 * np.log10(temp_psd[i]))
ax.set_xlim(0, 0.5)
ax.set_xlabel('Frequency ', fontsize=15)
ax.set_ylabel('PSD (dB)', fontsize=15)
plt.title('Преобразоваени Фурье', fontsize=25)

```

```python
from scipy import signal
import matplotlib.pyplot as plt
widths = np.arange(1, 61)
cwtmatr = signal.cwt(data_per, signal.ricker, widths)

fig = plt.figure(figsize=(18, 10))
ax.set_xlabel('(degC)')
plt.plot(np.sin(np.pi / 6))
plt.plot(cwtmatr[1])
plt.plot(cwtmatr[2])
plt.plot(cwtmatr[3])
plt.xlabel('Time', fontsize=15)
plt.ylabel('Amplitude', fontsize=15)
plt.title('Рассчитанные значения интегралов функции для каждого масштаба в каждый момент времени', fontsize=20)
```

```python
fig = plt.figure(figsize=(18, 10))
plt.title('Вейвлет преобразование Мексиканская шляпа', fontsize=20)
plt.imshow(cwtmatr, extent=[-1, 31,-1,1], cmap='PRGn', aspect='auto', vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max())
plt.xlabel('Time', fontsize=15)
plt.ylabel('Amplitude', fontsize=15)
plt.show()
```

```python
def DFT(x):
    """
    Compute the discrete Fourier Transform of the 1D array x
    :param x: (array)
    """

    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    return np.dot(e, x)
```

```python
import math
def polar(z):
    a= z.real
    b= z.imag
    r = math.hypot(a,b)
    theta = math.atan2(b,a)
    return r,theta # use return instead of print.

u=3+5j
print (polar(u))
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
 
# Скачиваем данные для графика.
url = 'https://python-scripts.com/wp-content/uploads/2019/3d-data.csv'
data = pd.read_csv(url)
 
# Преобразуем его в длинный формат
df = data.unstack().reset_index()
df.columns=["X", "Y", "Z"]
 
# Переименовываем старые названия столбцов в числовой формат.
df['X'] = pd.Categorical(df['X'])
df['X'] = df['X'].cat.codes
 
# Мы собираемся сделать 20 графиков, для 20 разных углов
for angle in range(70, 210, 2):
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    ax.plot_trisurf(df['Y'], df['X'], df['Z'], cmap=plt.cm.viridis, linewidth=0.2)
 
    ax.view_init(30, angle)
 
    filename = 'frames/step'+str(angle)+'.png'
    plt.savefig(filename, dpi=96)
    plt.gca()
```
