```python
# -*- coding: utf-8 -*-
import random
from collections import Counter
import scipy as sp
from scipy import stats
import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
def datat():
    return pd.read_excel('Temp.xlsx', encoding='windows-1252')
```

```python
def name():
    return datat().columns
name()
```

```python
def analiz1():
    return datat()['Temp']
analiz1()
```

```python
def wowdata():
    df = datat()
    return df[df['Temp'].notnull()]   
```

```python
def opisstat():
    return wowdata()['Temp'].count() 
opisstat()
```

```python
def mean(xs): 
    return sum(xs) / len(xs)
```

```python
def opisstat1():
    return wowdata()['Temp'].mean()
opisstat1()
```

```python
def opisstat2():
    return wowdata()['Temp'].median()

opisstat2()
```

```python
def variance(xs):
    x_hat = mean(xs)
    n = len(xs)
    n = n - 1 if n in range(1, 30) else n  
    square_deviation = lambda x : (x - x_hat) ** 2 
    return sum( map(square_deviation, xs) ) / n

def standard_deviation(xs):
    return sp.sqrt( variance(xs) )
```

```python
def opisstat3():
    return wowdata()['Temp'].std(ddof=0) 

opisstat3() 
```
